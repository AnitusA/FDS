```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
from sklearn.linear_model import LinearRegression 
from sklearn.model_selection import train_test_split 
from sklearn.metrics import r2_score, mean_squared_error 
# Step 1: Load dataset 
df = pd.read_csv("C:/Users/Anitus A/Downloads/fds/diabetes.csv") 
print("First 5 rows of dataset:\n", df.head()) 
# Step 2: Select independent (X) and dependent (y) variables 
X = df[['Pregnancies', 'Glucose', 'BloodPressure', 'Age']] 
y = df['BMI']   # Dependent variable: BMI 
# Step 3: Split into train and test sets 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, 
random_state=42) 
# Step 4: Fit multiple regression model 
model = LinearRegression() 
model.fit(X_train, y_train) 
# Step 5: Predictions 
y_pred = model.predict(X_test) 
# Step 6: Model evaluation 
print("Intercept:", model.intercept_) 
print("Coefficients:", model.coef_) 
print("R² Score:", r2_score(y_test, y_pred)) 
print("Mean Squared Error:", mean_squared_error(y_test, y_pred)) 
# Step 7: Plot actual vs predicted 
plt.scatter(y_test, y_pred, color='blue', alpha=0.6) 
plt.xlabel("Actual BMI") 
plt.ylabel("Predicted BMI") 
plt.title("Multiple Regression: Actual vs Predicted BMI") 
plt.show() 
```
