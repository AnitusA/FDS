```python
import numpy as np 
# Step 1: Create two arrays 
arr1 = np.array([10, -10, 10, -10, -10, 10]) 
arr2 = np.array([0.85, 0.45, 0.9, 0.8, 0.12, 0.6]) 
# Step 2: Define condition and count instances 
# Example: Count number of times value 10 appears in arr1 where arr2 > 0.5 
count = np.sum((arr1 == 10) & (arr2 > 0.5)) 
print("Number of instances of 10 in arr1 where arr2 > 0.5:", count)
```
