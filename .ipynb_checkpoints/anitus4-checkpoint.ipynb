```python
import numpy as np 
# Step 1: Create a 2D NumPy array of 4-byte integers 
arr = np.array([[1, 2, 3], 
[2, 1, 2]], dtype=np.int32) 
print("Original NumPy array:") 
print(arr) 
print("Type:", type(arr)) 
# Step 2: Define the sequence to search 
sequence = [2, 3] 
# Step 3: Convert array to a list of lists for easy sequence matching 
count = 0 
for row in arr: 
    for i in range(len(row) - len(sequence) + 1): 
        if list(row[i:i+len(sequence)]) == sequence: 
            count += 1 
            print("Sequence:", sequence) 
            print("Number of occurrences of the said sequence:", count) 
```
