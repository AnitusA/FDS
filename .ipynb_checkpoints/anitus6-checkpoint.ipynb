```python
import numpy as np 
# Step 1: Create two arrays 
arr1 = np.array(['PHP', 'JS', 'C++']) 
arr2 = np.array(['Python', 'C#', 'NumPy']) 
print("Original arrays:") 
print(arr1) 
print(arr2) 
# Step 2: Combine last element of arr1 with first element of arr2 
combined_element = arr1[-1] + arr2[0] 
# Step 3: Merge arrays with combined element 
merged_array = np.concatenate((arr1[:-1], [combined_element], arr2[1:])) 
# Step 4: Display result 
print("After Combining:") 
print(merged_array) 
```
