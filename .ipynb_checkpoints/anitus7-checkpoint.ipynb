```python
import pandas as pd 
import numpy as np 
# Step 1: Sample data dictionary 
data = {'X':[78,85,96,80,86], 
'Y':[84,94,89,83,86], 
'Z':[86,97,96,72,83]} 
# Step 2: Create DataFrame 
df = pd.DataFrame(data) 
# Step 3: Perform element-wise power (X^Y^Z row-wise) 
# Note: Since direct large powers will overflow, usually we demonstrate the logic. 
# Here, we will raise first column elements (X) to the power of second column (Y). 
result = np.power(df['X'], df['Y']) 
# Step 4: Display input and expected output 
print("Original DataFrame:") 
print(df) 
print("\nResult of element-wise power (X^Y):") 
print(result) 
```
