```python
import pandas as pd
import numpy as np
from scipy import stats

# Step 1: Load dataset
# Replace 'diabetes.csv' with actual path if running locally
df = pd.read_csv("C:/Users/Anitus A/Downloads/fds/diabetes.csv")

print("=== First 5 Rows of Dataset ===")
print(df.head(), "\n")

# Step 2: Frequency (for categorical variable 'Outcome')
print("=== Frequency of Outcome ===")
print(df['Outcome'].value_counts(), "\n")

# Step 3: Mean
print("=== Mean of each column ===")
print(df.mean(numeric_only=True), "\n")

# Step 4: Median
print("=== Median of each column ===")
print(df.median(numeric_only=True), "\n")

# Step 5: Mode

print("=== Mode of each column ===")
print(df.mode().iloc[0], "\n")

# Step 6: Variance
print("=== Variance of each column ===")
print(df.var(numeric_only=True), "\n")

# Step 7: Standard Deviation
print("=== Standard Deviation of each column ===")
print(df.std(numeric_only=True), "\n")

# Step 8: Skewness
print("=== Skewness of each column ===")
print(df.skew(numeric_only=True), "\n")

# Step 9: Kurtosis
print("=== Kurtosis of each column ===")
print(df.kurtosis(numeric_only=True), "\n")
```
