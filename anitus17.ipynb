```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D 

# Load dataset
df = sns.load_dataset('iris')

# Create 3D plot
fig = plt.figure(figsize=(7, 5))
ax = fig.add_subplot(111, projection='3d')

# Convert species to categorical codes
species = df['species'].astype('category')
codes = species.cat.codes  # 0,1,2

# Color mapping
cmap = plt.cm.viridis
norm = plt.Normalize(codes.min(), codes.max())

# Scatter plot
ax.scatter(
    df['sepal_length'],
    df['sepal_width'],
    df['petal_length'],
    c=cmap(norm(codes)),
    s=35, alpha=0.8
)

# Labels and title
ax.set_xlabel('Sepal Length')
ax.set_ylabel('Sepal Width')
ax.set_zlabel('Petal Length')
ax.set_title('3D Scatter: Sepal Length, Sepal Width, Petal Length')

handles = [
    Line2D([0], [0], marker='o', linestyle='',
           markerfacecolor=cmap(norm(i)), markeredgecolor='none',
           label=sp)
    for i, sp in enumerate(species.cat.categories)
]
ax.legend(handles=handles, title='Species', loc='upper left', bbox_to_anchor=(1.05, 1))

plt.tight_layout()
plt.show()
```
