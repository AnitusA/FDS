```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
from mpl_toolkits.mplot3d import Axes3D  # enables 3D plots 
# Step 1: Load the dataset 
df = pd.read_csv("C:/Users/Anitus A/Downloads/fds/diabetes.csv") 
# Step 2: Correlation and Scatter Plots 
plt.figure(figsize=(10,8)) 
sns.heatmap(df.corr(), annot=True, cmap="coolwarm", fmt=".2f") 
plt.title("Correlation Heatmap") 
plt.show() 
# Scatter plot example (Glucose vs BMI, colored by Outcome) 
plt.figure(figsize=(7,5)) 
sns.scatterplot(data=df, x="Glucose", y="BMI", hue="Outcome", alpha=0.7) 
plt.title("Scatter Plot: Glucose vs BMI by Outcome") 
plt.show() 
# Step 3: Histograms 
df.hist(figsize=(12,8), bins=20, edgecolor='black') 
plt.suptitle("Histograms of Diabetes Dataset Features") 
plt.show() 
# Histogram of Glucose separated by Outcome 
plt.figure(figsize=(7,5)) 
sns.histplot(data=df, x="Glucose", hue="Outcome", bins=20, kde=True, 
element="step") 
plt.title("Histogram of Glucose by Diabetes Outcome") 
plt.show() 
# Step 4: 3D Scatter Plot 
fig = plt.figure(figsize=(8,6)) 
ax = fig.add_subplot(111, projection="3d") 
colors = {0: "blue", 1: "orange"} 
for outcome, group in df.groupby("Outcome"): 
    ax.scatter(group["Glucose"], group["BMI"], group["Age"], 
c=colors[outcome], label=f"Outcome={outcome}", s=40, alpha=0.8) 
ax.set_xlabel("Glucose") 
ax.set_ylabel("BMI") 
ax.set_zlabel("Age") 
ax.set_title("3D Scatter Plot: Glucose, BMI, Age") 
ax.legend() 
plt.show() 
```
