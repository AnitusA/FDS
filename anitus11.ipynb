```python
import pandas as pd 
# Step 1: Create the DataFrame 
data = { 
'col1': ['C1', 'C1', 'C2', 'C2', 'C2', 'C3', 'C2'], 
'col2': [1, 2, 3, 3, 4, 6, 5] 
} 
df = pd.DataFrame(data) 
print("Original DataFrame") 
print(df) 
# Step 2: Group by col1 and aggregate col2 as lists 
result = df.groupby('col1')['col2'].apply(list) 
print("\nGroup on the col1:") 
print(result) 
```
